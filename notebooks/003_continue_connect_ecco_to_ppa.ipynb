```python
from ppanlp import *
pd.options.display.max_columns=None
def nozero(x):
    while x and x.startswith('0'):
        x=x[1:]
    return x

```

```python
dfmeta = pd.read_csv('./data/data.all_xml_metadata.ECCOII.csv')
dfmeta.iloc[0]
```

```python
df_excerpts = pd.read_csv('data/ECCO Excerpts-DigitizedWork-2023-09-20.csv')
# df_excerpts.iloc[0]
```

```python
df_excerpts_ecco2 = df_excerpts[df_excerpts.source_id.str.startswith('CB')]
df_excerpts_ecco2.iloc[0]
```

```python
# get the document IDs from the PPA ECCO2 excerpts file
ids = {x[2:] for x in df_excerpts_ecco2.source_id}
ids_no0 = {nozero(x) for x in ids}
```

```python
# get all viable types of doc ids from the metadata 
# retrieved from _DocMetadata.xml files

assetids = set(dfmeta['assetid'])
assetidstoc = set(dfmeta['assetidetoc'])
psmid = set(dfmeta['psmid'])
docids_from_links = {x.split('CB')[-1] for x in dfmeta['productlink']}
```

```python
# no matches in assetids
ids & assetids, ids_no0 & assetids
```

```python
# no matches
ids & assetidstoc, ids_no0 & assetidstoc
```

```python
# no matches
ids & psmid, ids_no0 & psmid
```

```python
# matches from Ids in links
ids & docids_from_links, ids_no0 & docids_from_links
```

```python
# no missing matches!
ids - docids_from_links
```

```python
# add the docid to the XML metadadta df
dfmeta['docid'] = dfmeta['productlink'].apply(lambda x: x.split('GALE%7C')[1])
dfmeta['docid']
```

```python
mergeddf = df_excerpts.merge(dfmeta, left_on='source_id', right_on='docid')
mergeddf
```

```python
mergeddf.to_csv('data/data.metadata.merged.ECCOII.csv', index=False)
```
